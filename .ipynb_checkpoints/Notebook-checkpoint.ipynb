# Model Building

```python
X = data.drop(['charges'], axis =1)
y = data.charges
```

```python
sex_mapping = {'female': 0, 'male': 1}

X['sex'] = X['sex'].map(sex_mapping)
```

```python
region_mapping = {
    'southwest': 0,
    'southeast': 1,
    'northwest': 2,
    'northeast': 3
}

X['region'] = X['region'].map(region_mapping)
```

```python
smoker_mapping = {'yes': 1, 'no': 0}


X['smoker'] = X['smoker'].map(smoker_mapping)
```

```python
X['region'].unique()
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python


def evaluate(model, X_train, X_test, y_train, y_test):
    # Training the model
    model.fit(X_train, y_train)
    
    # Predicting on training and testing data
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    print("TRAINING RESULTS: \n===============================")
    print(f"TRAINING R2 SCORE: {r2_score(y_train, y_train_pred):.4f}")
    print(f"TRAINING MEAN ABSOLUTE ERROR (MAE): {mean_absolute_error(y_train, y_train_pred):.4f}")
    print(f"TRAINING MEAN SQUARED ERROR (MSE): {mean_squared_error(y_train, y_train_pred):.4f}")
    print(f"TRAINING ROOT MEAN SQUARED ERROR (RMSE): {mean_squared_error(y_train, y_train_pred, squared=False):.4f}")

    print("\nTESTING RESULTS: \n===============================")
    print(f"TESTING R2 SCORE: {r2_score(y_test, y_test_pred):.4f}")
    print(f"TESTING MEAN ABSOLUTE ERROR (MAE): {mean_absolute_error(y_test, y_test_pred):.4f}")
    print(f"TESTING MEAN SQUARED ERROR (MSE): {mean_squared_error(y_test, y_test_pred):.4f}")
    print(f"TESTING ROOT MEAN SQUARED ERROR (RMSE): {mean_squared_error(y_test, y_test_pred, squared=False):.4f}")
```

# Multi Logistic Regression

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
start_time = time.time()
LR.fit(X_train, y_train)
end_time = time.time()
lr_time = end_time - start_time
```

```python
evaluate(LR, X_train, X_test, y_train, y_test)
```

```python
y_pred = LR.predict(X_test)

r2 = r2_score(y_test, y_pred)


mse = mean_squared_error(y_test, y_pred)


mae = mean_absolute_error(y_test, y_pred)


rmse = np.sqrt(mse)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()

start_time = time.time()

dt.fit(X_train, y_train)

end_time = time.time()

dt_time = end_time - start_time
```

```python
evaluate(dt, X_train, X_test, y_train, y_test)
```

```python
y_pred = dt.predict(X_test)

r21 = r2_score(y_test, y_pred)


mse1 = mean_squared_error(y_test, y_pred)


mae1 = mean_absolute_error(y_test, y_pred)


rmse1 = np.sqrt(mse)
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
start_time = time.time()
gb.fit(X_train, y_train)
end_time = time.time() 

gb_time = end_time - start_time
```

```python
evaluate(gb, X_train, X_test, y_train, y_test)
```

```python
y_pred = gb.predict(X_test)

r22 = r2_score(y_test, y_pred)


mse2 = mean_squared_error(y_test, y_pred)


mae2 = mean_absolute_error(y_test, y_pred)


rmse2 = np.sqrt(mse)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, random_state=0)
start_time = time.time()
rf.fit(X_train, y_train)
end_time = time.time() 

rf_time = end_time - start_time
```

```python
y_pred = rf.predict(X_test)

r23 = r2_score(y_test, y_pred)


mse3 = mean_squared_error(y_test, y_pred)


mae3 = mean_absolute_error(y_test, y_pred)


rmse3 = np.sqrt(mse)
```
